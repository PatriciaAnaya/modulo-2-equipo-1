El programa pedirá al usuario, mediante un input(), que escriba el nombre (o parte del nombre) de un personaje. A continuación, hará una búsqueda en el API de Star Wars y mostrará por consola un listado con los personajes que coinciden con la búsqueda, indicando su nombre y su género.

url base : https://swapi.py4e.com/api/

```python


personaje = input("Dime qué personaje quieres ver.")

url_personajes = f"https://swapi.py4e.com/api/people/?search={personaje}"
#para realizar una búsqueda dentro de la url, el símbolo "?" separa la url de la búsqueda.

personaje_escogido = requests.get(url_personajes)
personaje_json = personaje_escogido.json()
print(personaje_json)
print(personaje_json['name'], personaje_json['gender'])
```

```python
def personaje_sw():
    elegido = input("Dime qué personaje quieres ver")
    url_personajes = "https://swapi.py4e.com/api/people/"
    personaje_escogido = requests.get(url_personajes)
    personaje_json = personaje_escogido.json()

    for persona in personaje_json["results"]:
        if persona["name"] == elegido:
            print(persona['name'], persona["gender"])
```

```python
personaje_sw()
```

```python
url="https://api.balldontlie.io/v1/players"
api_key="e37eb514-eb3c-44df-90f4-9a496f776f16"
```

```python
nba_players=requests.get(url, headers={"Authorization": api_key} )
nba_players_json=nba_players.json()
nba_players.status_code
```

```python
nba_players_json
```

```python
jugadores_equipo=[]
for i in nba_players_json["data"]:
    print(i["first_name"], i["last_name"],i["team"]["full_name"])
    jugadores_equipo.append((i["first_name"], i["last_name"],i["team"]["full_name"]))
print(jugadores_equipo)
len(jugadores_equipo)
```

```python
jugadores=[]

def llamar_API():
    jugadores_A=requests.get(url,headers={"Authorization": api_key},params={"cursor":0,"per_page":100})
    jugadores_B=requests.get(url,headers={"Authorization": api_key},params={"cursor":100,"per_page":100})
    #jugadores.append(jugadores_A["data"])
    #jugadores.append(jugadores_B["data"])
    #jugadores_json=jugadores.json
    #jugadores_A["data"]
    jugadores_A_json=jugadores_A.json()
    jugadores_B_json=jugadores_B.json()
    #jugadores=jugadores_A_json | jugadores_B_json
    
    jugadores.extend(jugadores_A_json.get('data',[]))
    jugadores.extend(jugadores_B_json.get('data',[]))
    print(jugadores)

    return[jugadores]
```

```python
llamar_API()
```

```python
datos_jugadores=[]

for i in jugadores:
    print(i["first_name"], i["last_name"],i["team"]["full_name"])
    datos_jugadores.append((i["first_name"], i["last_name"],i["team"]["full_name"]))
print(datos_jugadores)
len(datos_jugadores)
```

```python
df_jugadores=pd.DataFrame(datos_jugadores, columns=["nombres ", "apellidos", "equipo"])
df_jugadores
```
