Buscar un personaje de Star Wars

Ahora que conocemos mejor el API de Star Wars, vamos a crear una aplicación en Python que la utilice. El programa pedirá al usuario, mediante un input(), que escriba el nombre (o parte del nombre) de un personaje. A continuación, hará una búsqueda en el API de Star Wars y mostrará por consola un listado con los personajes que coinciden con la búsqueda, indicando su nombre y su género.

url base : https://swapi.py4e.com/api/
autenticacion : no es necesario
método http _ http swapi.py4e.com/api/planets/1/
Luke : http https://swapi.py4e.com/api/people/1/
A New Hope : https://swapi.py4e.com/api/films/1
URL listado peronajes : https://swapi.py4e.com/api/people/
wookie : https://swapi.py4e.com/api/people/1/?format=wookiee


```python


personaje = input("Dime qué personaje quieres ver.")

url_personajes = f"https://swapi.py4e.com/api/people/?search={personaje}"
#para realizar una búsqueda dentro de la url, el símbolo "?" separa la url de la búsqueda.

personaje_escogido = requests.get(url_personajes)
personaje_json = personaje_escogido.json()
print(personaje_json)
print(personaje_json['name'], personaje_json['gender'])
```

```python
def personaje_sw():
    elegido = input("Dime qué personaje quieres ver")
    url_personajes = "https://swapi.py4e.com/api/people/"
    personaje_escogido = requests.get(url_personajes)
    personaje_json = personaje_escogido.json()

    for persona in personaje_json["results"]:
        if persona["name"] == elegido:
            print(persona['name'], persona["gender"])
```

```python
personaje_sw()
```
